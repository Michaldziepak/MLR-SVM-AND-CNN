```python
import pandas as pd
import matplotlib.pyplot as plt
Data = pd.read_excel("Linear_calc.xlsx")
Data
```

```python
cm = plt.cm.get_cmap('RdYlBu')
sc = plt.scatter(Data["testsize"], Data["valuelin"], c=Data["R^2scorelin"], vmin=0, vmax=1, s=35, cmap=cm)
plt.colorbar(sc)
plt.xlabel('Testsize')
plt.ylabel('Values')
plt.show()
```
